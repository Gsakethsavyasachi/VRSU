```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("/content/low pixel dog image.jpg",0)
he_img=cv2.equalizeHist(img,cv2.COLOR_BGR2GRAY)

#Display setting
plt.figure(figsize=(10,3))

plt.subplot(2,1,1)
plt.title("Original Image")
plt.imshow(img,cmap="gray")
plt.axis("off")

plt.subplot(2,1,2)
plt.title("Histogram Equalization")
plt.imshow(he_img,cmap="grey")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Graph")
plt.hist(he_img.ravel(),256)
plt.show()
```

# Spatial Domain Filtering - Smoothing

```python
import cv2
import matplotlib.pyplot as plt

img_avg=cv2.imread("/content/bricks.webp")
img_gauss=cv2.imread("/content/average blur.webp")
img_median=cv2.imread("/content/median blur.webp")

img_avg_rgb=cv2.cvtColor(img_avg,cv2.COLOR_BGR2RGB)
img_gauss_rgb=cv2.cvtColor(img_gauss,cv2.COLOR_BGR2RGB)
img_median_rgb=cv2.cvtColor(img_median,cv2.COLOR_RGB2BGR)

avg_blur=cv2.blur(img_avg_rgb,(5,5))
avg_blur_rgb=cv2.cvtColor(avg_blur,cv2.COLOR_BGR2RGB)

gauss_blur=cv2.GaussianBlur(img_gauss_rgb,(5,5),0)
gauss_blur_rgb=cv2.cvtColor(gauss_blur,cv2.COLOR_BGR2RGB)

median_blur=cv2.medianBlur(img_median_rgb,5)
median_blur_rgb=cv2.cvtColor(median_blur,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(14,10))

#Average Blur
plt.subplot(3,2,1)
plt.imshow(img_avg_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(3,2,2)
plt.imshow(avg_blur)
plt.title("Average Blur")
plt.axis("off")

#Gaussian Blur
plt.subplot(3,2,3)
plt.imshow(img_gauss_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(3,2,4)
plt.imshow(gauss_blur)
plt.title("Gaussian Blur")
plt.axis("off")

#Median Blur
plt.subplot(3,2,5)
plt.imshow(img_median_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(3,2,6)
plt.imshow(median_blur)
plt.title("Median Blur")
plt.axis("off")

plt.show()
```
